Plotting
matplotlib.pyplot has a lot of options. Here are a couple of useful ways to make figures.

```python
# plt.plot plots lines or points, with basic colours
xaxis=np.arange(16)*(np.pi/4)
plt.plot(xaxis,np.sin(xaxis),'rx')

# the inputs are x values, y values, and plotting parameters
# available colours are k, r, g, b, c, m, y
# available point styles are ., o, ^, s, x
```

```python
# you can also control the scale of the axes and add labels (with tex markup)
xaxis=np.arange(16)*(np.pi/4)
plt.plot(xaxis,np.sin(xaxis),'ko')
plt.xlim(-5,50)
plt.ylim(2,-3)
plt.xlabel('$\phi$')
plt.ylabel('sin($\phi$)')
plt.title('pyplot example')
```

```python
# You can also connect the points with a line
plt.plot(xaxis,np.sin(xaxis),'b--')

# line styles include -, --, -., 
```

```python
# plt.scatter is a bit more configurable
plt.scatter(xaxis,(xaxis**2)*np.exp(-xaxis/3),s=30,c='seagreen')

# s sets the point size
# c sets the colour: options are listed at https://matplotlib.org/gallery/color/named_colors.html
```

```python
# you can also set the colour to have values from the data
# and you can choose a colourmap
plt.scatter(xaxis,(xaxis**2)*np.exp(-xaxis/3),s=30,c=np.sqrt(xaxis),cmap='cool')

# cmap options are listed at https://matplotlib.org/examples/color/colormaps_reference.html
```

Lets do a linear fit

```python
# Create some linear data
np.random.seed(0)
x = np.linspace(0, 10, 50)
y = 3 * x + 2 + np.random.normal(0, 2, size=x.shape)
```

```python
# Perform linear regression using scipy.stats.linregress
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
```

```python
# Print regression results
regression_results = {
    'slope': slope,
    'intercept': intercept,
    'r_value': r_value,
    'p_value': p_value,
    'std_err': std_err
}

# Print the regression results
print("Regression Results:")
for key, value in regression_results.items():
    print(f"{key}: {value}")
```

```python
# Plotting the data and the fitted line
plt.scatter(x, y, label="Data")
plt.plot(x, slope * x + intercept, color='red', label="Fitted Line")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
